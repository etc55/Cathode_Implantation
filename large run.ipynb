```python
df
```

```python
d = {'eventID': df[2][1:], 'edep': df[3][1:], 'z0': df[10][1:], 'typ': df[36][1:], 't0': df[6][1:], 'trackID': df[33][1:], 'stepnum':df[34][1:], 'parent': df[18][1:]}
df2 = pd.DataFrame(data=d)
df2['eventID'] = pd.to_numeric(df2['eventID'])
df2['trackID'] = pd.to_numeric(df2['trackID'])
df2['edep']=pd.to_numeric(df2['edep'])
df2['z0'] = pd.to_numeric(df2['z0'])
df2['stepnum'] = pd.to_numeric(df2['stepnum'])

dfpo = df2 [df2['parent'] != 'Bi214']
dfposort= dfpo.sort_values(['eventID','trackID','stepnum'], ascending=True)
dfposort
```

```python
df[18]
```

```python
event = []
layer = []
energy = []
typ = []
par = []
edep = []

evt = 0
lyr = 0
t0 = 0

for index, row in dfposort.iterrows():
    if float(row['eventID']) != evt:
        evt = (row['eventID'])
        lyr2= (row['z0'])
        lyr = -(lyr2-.06366)*10000000//width
        
    event.append(evt)
    layer.append(lyr)
    energy.append((row['edep']))
    typ.append(row['typ'])
    par.append(row['parent'])
    edep.append(row['edep'])
    
d2={'event': event, 'layer': layer, 'energy': energy, 'type': typ, 'parent': par, 'edep': edep}
df3=pd.DataFrame(data=d2)

df4=df3[df3['layer'] < 121]
df4=df4[df4['layer'] >= 0]

    
df4
```

```python
dweights = []
nrgs = []

for index, row in df4.iterrows():
    dweights.append(bars[int(row['layer'])])
    nrgs.append(row['edep'])
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter as pf

n = sum(dweights)
h=plt.hist(nrgs, bins=500, weights=dweights/n)
plt.xlabel("energy deposited (keV)")
plt.ylabel("percentage of events")
plt.title("my test plot")
plt.gca().yaxis.set_major_formatter(pf(1))
plt.show
```
